## Companies Revenue Data on Wikepedia

```python
url_companies = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page_request = requests.get(url_companies)

soup2 = BeautifulSoup(page_request.text, 'html')
```

```python
soup2.find_all("th")
```

```python
soup2.find_all("table")[1]
```

```python
soup2.find_all("table", class_ ="wikitable sortable")
```

```python
table = soup2.find_all("table")[1]
```

```python
table
```

```python
titles = table.find_all("th")
```

```python
titles
```

```python
#we are going to try and loop through all the titles and extract only the text from each one of them
#strip cleans each of the titles up by removing spaces 

world_table_titles = [title.text.strip() for title in titles]
```

```python
world_table_titles
```

```python
companies_df = pd.DataFrame(columns=world_table_titles)
```

```python
companies_df 
```

```python
rows = table.find_all("tr")


for row in rows[1:]: 
        row_data = row.find_all("td")
        individual_data = [data.text.strip() for data in row_data]
        length =len(companies_df)
        companies_df.loc[length]=individual_data 
        #we add this row to the next position in the dataframe
       
```
